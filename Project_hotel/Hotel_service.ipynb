```python
#создаем датасет
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Функция для генерации случайной даты
def random_date(start, end):
    return start + timedelta(days=random.randint(0, (end - start).days))

# Настройки генерации данных
num_rows = 1000
hotels = ['City Hotel', 'Resort Hotel']
meal_options = ['BB', 'HB', 'FB', 'SC', 'Undefined']
countries = ['USA', 'UK', 'Germany', 'France', 'Italy', 'Spain', 'Brazil', 'Russia', 'China', 'India']
room_types = ['A', 'B', 'C', 'D', 'E']
customer_types = ['Transient', 'Contract', 'Group', 'Transient-Party']
reservation_statuses = ['Canceled', 'Check-Out', 'No-Show']

# Генерация данных
data = {
    'Hotel': [random.choice(hotels) for _ in range(num_rows)],
    'Is canceled': [random.choice([0, 1]) for _ in range(num_rows)],
    'Lead time': [random.randint(1, 365) for _ in range(num_rows)],
    'Arrival full date': [random_date(datetime(2020, 1, 1), datetime(2023, 12, 31)).strftime("%Y-%m-%d") for _ in range(num_rows)],
    'Arrival date year': [random_date(datetime(2020, 1, 1), datetime(2023, 12, 31)).year for _ in range(num_rows)],
    'Arrival date month': [random_date(datetime(2020, 1, 1), datetime(2023, 12, 31)).month for _ in range(num_rows)],
    'Arrival date week number': [random_date(datetime(2020, 1, 1), datetime(2023, 12, 31)).isocalendar()[1] for _ in range(num_rows)],
    'Arrival date day of month': [random_date(datetime(2020, 1, 1), datetime(2023, 12, 31)).day for _ in range(num_rows)],
    'Stays in weekend nights': [random.randint(0, 5) for _ in range(num_rows)],
    'Stays in week nights': [random.randint(1, 14) for _ in range(num_rows)],
    'Stays total nights': [0]*num_rows,  # Placeholder to calculate later
    'Adults': [random.randint(1, 4) for _ in range(num_rows)],
    'Children': [random.randint(0, 3) for _ in range(num_rows)],
    'Babies': [random.randint(0, 1) for _ in range(num_rows)],
    'Meal': [random.choice(meal_options) for _ in range(num_rows)],
    'Country': [random.choice(countries) for _ in range(num_rows)],
    'Reserved room type': [random.choice(room_types) for _ in range(num_rows)],
    'Assigned room type': [random.choice(room_types) for _ in range(num_rows)],
    'Customer type': [random.choice(customer_types) for _ in range(num_rows)],
    'Reservation status': [random.choice(reservation_statuses) for _ in range(num_rows)],
    'Reservation status date': [random_date(datetime(2020, 1, 1), datetime(2023, 12, 31)).strftime("%Y-%m-%d") for _ in range(num_rows)],
}

# Создание DataFrame
df = pd.DataFrame(data)

# Расчет общего числа ночей
df['Stays total nights'] = df['Stays in weekend nights'] + df['Stays in week nights']

# Сохранение в CSV файл 
df.to_csv('C:/Users/majkl/WorkSpace/Project_hotel/hotel_bookings_dataset.csv', index=False)


print(df.head())
```

Не все бронирования завершились успешно (is_canceled), поэтому можно посчитать, сколько клиентов было потеряно в процессе. Иными словами, посчитать метрику под названием Churn Rate.

Churn rate (отток, коэффициент оттока) – это процент подписчиков (например, на push-уведомления от сайта), которые отписались от канала коммуникации, отказались от услуг сервиса в течение определенного периода времени. Иными словами, представляет собой отношение количества ушедших пользователей к общему количеству пользователей, выраженное в процентах.

В нашем случае Churn Rate - это процент клиентов, которые отменили бронированКм, как эта метрика связана с наличием детей у клиентов!
Создайте колонку has_kids, которая принимает значение True, если клиент при бронировании указал хотя бы одного ребенка (total_kids), в противном случае – False. Далее проверьте, среди какой группы пользователей показатель оттока выше. 


```python
def has_kids(x):
    if x > 0:
        return True
    else:
        return False
```

```python
df['has_kids'] = df.total_kids.apply(has_kids)
```

```python
df['has_kids'].value_counts()
```

```python
df[['has_kids', 'is_canceled']].value_counts() / df[['has_kids']].value_counts() * 100
```
